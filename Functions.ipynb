### Q2. Explain the concept of function arguments and parameters in Python.

**Answer:**

In Python, **parameters** and **arguments** are related to functions.

- **Parameters** are the variables listed inside the parentheses in the function definition.
- **Arguments** are the actual values passed to the function when it is called.

When a function is defined, parameters act as placeholders. When the function is called, the arguments replace these placeholders with real values.

**Example:**

```python
def greet(name):  # 'name' is a parameter
    print("Hello", name)

greet("Palak")     # "Palak" is an argument

### Q6. Explain the concept of generators in Python and how they are defined.

**Answer:**

Generators are special functions that return an iterator using the `yield` keyword. They produce values one at a time and are memory efficient.

**Example:**
```python
def countdown():
    yield 3
    yield 2
    yield 1

for num in countdown():
    print(num)
```

### Q7. What are the advantages of using generators over regular functions?

**Answer:**

- Generators use less memory by yielding one value at a time.
- They are suitable for large datasets or infinite sequences.
- They save state between iterations automatically.

**Example:**
```python
def infinite_numbers():
    n = 0
    while True:
        yield n
        n += 1
```

### Q9. Explain the purpose and usage of the `map()` function in Python.

**Answer:**

The `map()` function applies a given function to each item of an iterable (like a list) and returns a map object.

**Example:**
```python
def square(x):
    return x * x

nums = [1, 2, 3, 4]
squares = list(map(square, nums))
print(squares)  # [1, 4, 9, 16]
```

### Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

**Answer:**

| Function | Purpose |
|---------|---------|
| `map()` | Applies a function to each element. |
| `filter()` | Filters elements where the function returns True. |
| `reduce()` | Applies a function cumulatively to the items. |

**Example:**
```python
from functools import reduce

nums = [1, 2, 3, 4, 5]

squares = list(map(lambda x: x*x, nums))
evens = list(filter(lambda x: x % 2 == 0, nums))
total = reduce(lambda x, y: x + y, nums)
```

````python
### Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

print(sum_even_numbers([1, 2, 3, 4, 5, 6]))
````

````python
### Q3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_numbers(numbers):
    return [num**2 for num in numbers]

print(square_numbers([1, 2, 3, 4]))
````

````python
### Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

print([num for num in range(1, 201) if is_prime(num)])
````

````python
### Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

for num in Fibonacci(10):
    print(num, end='')
````

````python
### Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for val in powers_of_two(5):
    print(val)
````

````python
### Q7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as f:
        for line in f:
            yield line.strip()

with open("sample.txt", "w") as f:
    f.write("Line 1\nLine 2\nLine 3")

for line in read_file_line_by_line("sample.txt"):
    print(line)
````

````python
### Q9. Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.
celsius = [0, 20, 37, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
````

````python
### Q10. Create a Python program that uses filter() to remove all the vowels from a given string.
def remove_vowels(string):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda ch: ch not in vowels, string))

print(remove_vowels("Hello World"))
````

````python
# Q11. Accounting routine using lambda and map
# We are given a list of orders from a book shop.
# Each order is a list containing:
# 1. Order number
# 2. Book title and author
# 3. Quantity
# 4. Price per item
#
# Requirement:
# Calculate total price for each order.
# If total < 100, add 10 as a surcharge.
# Return a list of tuples: (order number, total amount)

orders = [
    [34587, 'Learning Python, Mark Lutz', 4, 40.95],
    [98762, 'Programming Python, Mark Lutz', 5, 56.80],
    [77226, 'Head First Python, Paul Barry', 3, 32.95],
    [88112, 'Einführung in Python3, Bernd Klein', 3, 24.99]
]

# Using lambda and map
final_result = list(map(
    lambda order: (
        order[0],
        order[2] * order[3] + 10 if order[2] * order[3] < 100 else order[2] * order[3]
    ),
    orders
))

print(final_result)
````
